```python
#ID no: 18IT033
#Assigning column names for the dataset
headernames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
#importing dataset
dataset = pd.read_csv('iris.data', names = headernames)
dataset.head()
```

```python
#Total observations in the dataset
print("Count of samples in each class before encoding:\n",dataset['Class'].value_counts())
print("Classes:\n",dataset['Class'])
print(dataset.shape)
print(dataset.count())
```

# **Encoding**

```python
# Importing LabelEncoder and initializing it
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Class']=le.fit_transform(dataset['Class'])
print("Count of samples in each class after encoding:\n",dataset['Class'].value_counts())
print("Classes: ",le.classes_)
dataset
```

```python
#Performing one hot encoding
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(dataset['Class'].values.reshape(-1,1)).toarray()
print("Encoded Categories: ",encoder.categories_)


#Transforming the encoded data to dataframe
encoded_data = pd.DataFrame(encoded_data , columns = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
encoded_data
```

# **Standardization**

```python
#Standardization
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X = dataset.drop(['Class'], axis = 'columns')
y = dataset.Class
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

#Performing scaling using StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print("Scaled train data: ", X_train)
```

# **Normalization**

```python
#Normalization
from sklearn.preprocessing import normalize
X = dataset.drop(['Class'], axis = 'columns')
X_normalized = normalize(X, norm='l2')
print("Normalized Data:\n",X_normalized)
```

# **Discretization**

```python
#Discretization
from sklearn.preprocessing import KBinsDiscretizer
kbins = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
X = dataset.drop(['Class'], axis = 'columns')
data_trans = kbins.fit_transform(X)
print("Data after performing discretization:\n",data_trans)
```
